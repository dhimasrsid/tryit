```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl
```

```python
data = pd.read_excel('D:\dahmas\Tugas Ngampus\smester3\Employee Sample Data.xlsx')
data
```

```python
avg_age = data.groupby(['Gender', 'Department'])['Age'].mean().reset_index()

# Plot lineplot
sns.lineplot(
    x="Department",
    y="Age",
    hue="Gender",
    style="Gender",
    data=avg_age,
    marker="o"
)

plt.title('Rata-rata Umur Karyawan per Department dan Gender')
plt.xlabel('Department')
plt.ylabel('Rata-rata Umur')
plt.xticks(rotation=45)
plt.show()
```
